### Imports

```python
import cv2 
import os, glob

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

image_dir = "images/hand_gestures_digits_0_5"
```

### Plot all the base images in a subplot of 2 rows and 3 columns

```python
## Code to plot the base images.
img0 = cv2.imread('images/hand_gestures_digits_0_5/01.jpg')
img0 = cv2.cvtColor(img0, cv2.COLOR_BGR2RGB)
img0 = cv2.resize(img0, (400, 500))

img1 = cv2.imread('images/hand_gestures_digits_0_5/02.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img1 = cv2.resize(img1, (400, 500))

img2 = cv2.imread('images/hand_gestures_digits_0_5/03.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2, (400, 500))

img3 = cv2.imread('images/hand_gestures_digits_0_5/04.jpg')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img3 = cv2.resize(img3, (400, 500))

img4 = cv2.imread('images/hand_gestures_digits_0_5/05.jpg')
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
img4 = cv2.resize(img4, (400, 500))

img5 = cv2.imread('images/hand_gestures_digits_0_5/06.jpg')
img5 = cv2.cvtColor(img5, cv2.COLOR_BGR2RGB)
img5 = cv2.resize(img5, (400, 500))

f, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(20,10))
ax1.set_title("0")
ax1.imshow(img0)
ax2.set_title("1")
ax2.imshow(img1)
ax3.set_title("2")
ax3.imshow(img2)
ax4.set_title("3")
ax4.imshow(img3)
ax5.set_title("4")
ax5.imshow(img4)
ax6.set_title("5")
ax6.imshow(img5)
```

### Process all the images. Add all the code cells you may require.

```python
## Image processing.

# Convert to grayscale for filtering
gray_img0 = cv2.cvtColor(img0, cv2.COLOR_RGB2GRAY)
gray_img1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
gray_img3 = cv2.cvtColor(img3, cv2.COLOR_RGB2GRAY)
gray_img4 = cv2.cvtColor(img4, cv2.COLOR_RGB2GRAY)
gray_img5 = cv2.cvtColor(img5, cv2.COLOR_RGB2GRAY)

# blur background
gray_img0_blur = cv2.GaussianBlur(gray_img0, (5,5), 0)
gray_img1_blur = cv2.GaussianBlur(gray_img1, (3,3), 0)
gray_img2_blur = cv2.GaussianBlur(gray_img2, (3,3), 0)
gray_img3_blur = cv2.GaussianBlur(gray_img3, (3,3), 0)
gray_img4_blur = cv2.GaussianBlur(gray_img4, (3,3), 0)
gray_img5_blur = cv2.GaussianBlur(gray_img5, (3,3), 0)

# define lower and upper thresholds for hysteresis
lower = 10
upper = 100

img0_edges = cv2.Canny(gray_img0_blur, lower, upper)
img1_edges = cv2.Canny(gray_img1_blur, lower, upper)
img2_edges = cv2.Canny(gray_img2_blur, lower, upper)
img3_edges = cv2.Canny(gray_img3_blur, lower, upper)
img4_edges = cv2.Canny(gray_img4_blur, lower, upper)
img5_edges = cv2.Canny(gray_img5_blur, lower, upper)

# morphological operations

kernel = np.ones((5,5), np.uint8)

img0_dilation = cv2.dilate(img0_edges, kernel, iterations = 1)
img1_dilation = cv2.dilate(img1_edges, kernel, iterations = 1)
img2_dilation = cv2.dilate(img2_edges, kernel, iterations = 1)
img3_dilation = cv2.dilate(img3_edges, kernel, iterations = 1)
img4_dilation = cv2.dilate(img4_edges, kernel, iterations = 1)
img5_dilation = cv2.dilate(img5_edges, kernel, iterations = 1)
```

### Plot all the processed images in a subplot of 2 rows and 3 columns.

```python
## Code to plot the processed images.
f, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(20,10))
ax1.set_title("0")
ax1.imshow(img0_dilation, cmap="gray")
ax2.set_title("1")
ax2.imshow(img1_dilation, cmap="gray")
ax3.set_title("2")
ax3.imshow(img2_dilation, cmap="gray")
ax4.set_title("3")
ax4.imshow(img3_dilation, cmap="gray")
ax5.set_title("4")
ax5.imshow(img4_dilation, cmap="gray")
ax6.set_title("5")
ax6.imshow(img5_dilation, cmap="gray")
```
